```python
import pandas as pd

# Dataset
data = {
    "Team": ["MI", "CSK", "RCB", "MI", "CSK", "RCB"],
    "Player": ["Rohit", "Dhoni", "Kohli", "Sky", "Jadeja", "Maxwell"],
    "Runs": [70, 45, 90, 55, 30, 20],
    "Wickets": [0, 1, 0, 0, 2, 1]
}
ipl = pd.DataFrame(data)

# Total runs scored by each team
team_runs = ipl.groupby("Team")["Runs"].sum()
print(team_runs)
```

```python
# 2. Identify the highest run scorer from every team.
highest_scorers = ipl.loc[ipl.groupby("Team")["Runs"].idxmax()]
print(highest_scorers[["Team", "Player", "Runs"]])
```

```python
#3. Count how many players belong to each team.
# Count players per team
player_count = ipl.groupby("Team")["Player"].count()
print(player_count)
```

```python
#Compute the average number of wickets taken by each team
avg_wickets = ipl.groupby("Team")["Wickets"].mean()
print(avg_wickets)
```

```python
#Filter players who scored more than 50 runs
players_above_50 = ipl[ipl["Runs"] > 50]
print(players_above_50)



```

```python
#  Display all players belonging to MI or CSK
mi_csk_players = ipl[ipl["Team"].isin(["MI", "CSK"])]
print(mi_csk_players)
```

```python
# Extract players with either zero wickets OR more than 60 runs
filtered_players = ipl[(ipl["Wickets"] == 0) | (ipl["Runs"] > 60)]
print(filtered_players)
```

```python
team_runs = ipl.groupby("Team")["Runs"].sum()
teams_above_100 = team_runs[team_runs > 100]
print(teams_above_100)
```

```python
team_stats = ipl.groupby("Team")["Runs"].agg(["mean", "max"])
print(team_stats)
```

```python
rcb_below_50 = ipl[(ipl["Team"] == "RCB") & (ipl["Runs"] < 50)]
print(rcb_below_50)
```

# Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Create IPL dataset
data = {
    "Team": ["MI", "CSK", "RCB", "MI", "CSK", "RCB"],
    "Player": ["Rohit", "Dhoni", "Kohli", "Sky", "Jadeja", "Maxwell"],
    "Runs": [70, 45, 90, 55, 30, 20],
    "Wickets": [0, 1, 0, 0, 2, 1]
}
ipl = pd.DataFrame(data)

# Plot bar chart of runs by each player
plt.figure(figsize=(8,6))
plt.bar(ipl["Player"], ipl["Runs"], color="skyblue", edgecolor="black")

# Add labels and title
plt.xlabel("Players")
plt.ylabel("Runs Scored")
plt.title("Runs Scored by Each Player in IPL Dataset")
plt.show()
```

```python
# Calculate total runs per team
team_runs = ipl.groupby("Team")["Runs"].sum()

# Plot bar chart
plt.figure(figsize=(8,6))
team_runs.plot(kind="bar", color="orange", edgecolor="black")

# Add labels and title
plt.xlabel("Teams")
plt.ylabel("Total Runs")
plt.title("Total Runs Scored by Each Team")
plt.xticks(rotation=0)
plt.show()
```

```python
# Line plot of players vs runs
plt.figure(figsize=(8,6))
plt.plot(ipl["Player"], ipl["Runs"], marker="s", linestyle="-", color="green")

# Add labels and title
plt.xlabel("Players")
plt.ylabel("Runs Scored")
plt.title("Line Plot of Players vs Runs Scored")
plt.grid(True)
plt.show()
```

```python
# Horizontal bar chart
plt.figure(figsize=(8,6))
plt.barh(ipl["Player"], ipl["Wickets"], color="purple")
plt.xlabel("Wickets Taken")
plt.ylabel("Players")
plt.title("Wickets Taken by Each Player")
plt.show()
```

```python
team_counts = ipl["Team"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(team_counts, labels=team_counts.index, autopct="%1.1f%%", startangle=90, colors=["gold","lightblue","lightgreen"])
plt.title("Number of Players per Team")
plt.show()
```

```python
# Scatter plot of Runs vs Wickets with team color coding
plt.figure(figsize=(8,6))
sns.scatterplot(data=ipl, x="Runs", y="Wickets", hue="Team", s=100, palette="Set2")
plt.title("Scatter Plot of Runs vs Wickets (Team Color Coded)")
plt.show()
```

```python
# Average runs by team using a Seaborn barplot
plt.figure(figsize=(8,6))
sns.barplot(data=ipl, x="Team", y="Runs", hue="Team", estimator="mean", palette="coolwarm", legend=False)
plt.title("Average Runs by Team")
plt.show()
```

```python
# Countplot of the number of players per team
plt.figure(figsize=(8,6))
sns.countplot(data=ipl, x="Team", hue="Team", palette="muted", legend=False)
plt.title("Number of Players per Team")
plt.show()
```

```python
# Correlation heatmap for Runs and Wickets
plt.figure(figsize=(6,4))
corr = ipl[["Runs","Wickets"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap: Runs vs Wickets")
plt.show()

```
